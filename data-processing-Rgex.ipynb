```python
pattern = 'hello'
result = re.match(pattern=pattern,string=text)
```

```python
result.group()
```

```python
pattern = r'python'
result = re.search(pattern=pattern,string=text)
```

```python
print(result)
```

```python
pattern = r'\w+@\w+.\w+'
result = re.search(pattern=pattern,string=text)
```
